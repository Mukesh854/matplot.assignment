Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

A scatter plot is a type of data visualization that uses dots to represent individual data points on a two-dimensional plane.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot ")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib's Pyplot module is used to create a grid of subplots within a single figure.

```python
data = [
    (np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500]), 'Line 1'),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70]), 'Line 2'),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60]), 'Line 3'),
    (np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]), 'Line 4')
]


plt.figure(figsize=(10, 6)) 

for i, (x, y, label) in enumerate(data, start=1):
    plt.subplot(2, 2, i)
    plt.plot(x, y, marker='o', linestyle='-', label=label)
    plt.title(f'Line Plot {i}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()


plt.tight_layout()


plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

A bar plot, also known as a bar chart, is a data visualization that represents categorical data using rectangular bars.Bar plots are commonly used for the comparison , distribution and frequency count.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# vertical bar plot"
plt.figure(figsize=(10,6))
plt.bar(company,profit,color="red")
plt.xlabel("company")
plt.ylabel("profit")
plt.title("vertical bar plot")
plt.show()
```

```python
#horizontal bar plot

plt.figure(figsize=(10,6))
plt.barh(company,profit,color="red")
plt.xlabel("profit")
plt.ylabel("company")
plt.title("horizontal bar plot")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a data visualization that provides a summary of the distribution of a dataset.Box plots are commonly used for the following comparison, distribution , identification of outlier , skewness and symmetry.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.figure(figsize=(8,4))
plt.boxplot([box1,box2],vert=True,labels=["Box1", "Box2"])
plt.xlabel("Boxes")
plt.ylabel("Values")
plt.title("Box plot")
plt.show()
```
